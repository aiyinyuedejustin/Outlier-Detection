```python
import  matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from pyod.models.hbos import HBOS
from pyod.models.pca import PCA
from time import time


df_time = pd.read_csv("TimeData.csv")
df_precision = pd.read_csv("PrecisionData.csv")
df_recall = pd.read_csv("RecallData.csv")
```

```python
df_time
```

```python
# 1.	Time complixty, different sample size, just on testing size 0.2 for all 10 algorithms
# 2.	Time complixty, different testing size, just on sample size 5000 for all 10 algorithms

# 3.	Roc    different sample size, just on testing size 0.2 for all 10 algorithms
# 4.	Roc    different testing size, just on sample size 5000 for all 10 algorithms

# 5.	Presision different sample size, just on testing size 0.2 for all 10 algorithms
# 6.	Presision different testing size, just on sample size 5000for all 10 algorithms
```

```python
df_time_graph_testing_size20 = df_time[df_time['Testing Size']==0.2]
df_time_graph_sample_size5000 = df_time[df_time['Data Sample']==5000]

```

```python
df_precision_graph_testing_size20 = df_precision[df_precision['Testing Size']==0.2]
df_precision_graph_sample_size5000 = df_precision[df_precision['Data Sample']==5000]

```

```python
df_recall_graph_testing_size20 = df_recall[df_recall['Testing Size']==0.2]
df_recall_graph_sample_size5000 = df_recall[df_recall['Data Sample']==5000]

```

```python
columns = ['ABOD', 'CBLOF', 'FB',
       'HBOS', 'IForest', 'KNN', 'LOF', 'MCD', 'OCSVM', 'PCA']
len(columns)
```

# This is execution Time Area

# this is when Test size 20

```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
 
x = df_time_graph_testing_size20['Data Sample']

y1= df_time_graph_testing_size20['ABOD']
plt.plot(x,y1)

y2 = df_time_graph_testing_size20["CBLOF"]
plt.plot(x,y2)

y1= df_time_graph_testing_size20['FB']
plt.plot(x,y1)

y2 = df_time_graph_testing_size20['HBOS']
plt.plot(x,y2)



y1= df_time_graph_testing_size20['IForest']
plt.plot(x,y1)

y2 = df_time_graph_testing_size20['KNN']
plt.plot(x,y2)


y1= df_time_graph_testing_size20['LOF']
plt.plot(x,y1)

y2 = df_time_graph_testing_size20['MCD']
plt.plot(x,y2)


y1= df_time_graph_testing_size20['OCSVM']
plt.plot(x,y1)

y2 = df_time_graph_testing_size20['PCA']
plt.plot(x,y2)


plt.xlabel("Data Sample")
plt.ylabel("Execution Time")


plt.legend(columns)

plt.show()
    
    
    
    
```
